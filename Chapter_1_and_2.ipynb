## Introduction

Neural Networks are a type of machine learning and are often used for "deep learning". What makes a neural network "deep" is when two or more hidden layers are used in the network. Most nerual networks used today are "deep".

Neural networks were conceived in the 1940s. We were only able to train them correctly when backpropoagation was proposed in the 1960s. However, only with the rise of computational power and speed were neural networks able to gain traction. In 2010, neural networks were at the forefront of machine learning in image captioning, language translation, audio and visual synthesis, self-driving cars, calculating risk, detecting fraud and cancer diagnosis etc.

## What is a Neural Network

Neural networks are inspired by the brain. They consist of functional units called neurons. Each neuron has inputs (dendrites) that are weighted and then summed, a bias for the inputs, an activation function and output (axon terminals). Neurons are arranged in layers. The outputs of a preceding layer become the inputs for every neuron in the next layer. The weights and biases for inputs are "trainable". Meaning they are adjusted during the learning process, this is performed through back propogation.


## The effect of weights and bias

```python
import numpy as np
import pandas as pd
from scipy.special import expit

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(9, 6))
plt.title("The effect of weights and biases")
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(x=0, c="grey")
plt.axhline(y=0, c="grey",)

weight = 1
bias = 0

X = np.arange(-3, 3.5, 0.5)
Y = X*weight+bias

plt.plot(X, Y, c='red', label="weight = 1, bias = 0")

weight = 2
bias = 0

Y2 = X*weight+bias

plt.plot(X, Y2, c='blue', label="weight = 2, bias = 0")

weight = 1
bias = 2

Y3 = X*weight+bias

plt.plot(X, Y3, c='green', label="weight = 1, bias = 2")

plt.legend()
```

## The effect of activation functions

```python
plt.figure(figsize=(9, 6))
plt.title("The effect of activation functions")
plt.xlabel("Y")
plt.ylabel("Y_activation")
plt.axvline(x=0, c="grey")
plt.axhline(y=0, c="grey",)

weight = 1
bias = 0

Y = np.arange(-3, 3.5, 0.001)
Y_activation = np.empty_like(Y)


def step_function(Y):
    if Y < 0:
        Y_activation = 0
    elif Y >= 0:
        Y_activation = 1
    return Y_activation


pos = 0
for y in Y:
    y_activation = step_function(y)
    Y_activation[pos] = y_activation
    pos += 1


plt.plot(Y, Y_activation, c='red', label="step activation")

Y_activation = np.empty_like(Y)

pos = 0
for y in Y:
    y_activation = expit(y)
    Y_activation[pos] = y_activation
    pos += 1

plt.plot(Y, Y_activation, c='blue', label="sigmoid activation")

Y_activation = np.empty_like(Y)


def relu(x):
    return max(0.0, x)


pos = 0
for y in Y:
    y_activation = relu(y)
    Y_activation[pos] = y_activation
    pos += 1

plt.plot(Y, Y_activation, c='green', label="relu activation")

plt.legend()
```

## Matrix product

Since batches are conveniant, the dot products of 2 matrices will result in a matrix of results. Essentially a list of dots products. The shape of the matrices need to be compatible. There needs to be the same dimension on the rows of one matrix vs the columns in the other. The example below is compatible because the number of columns in X is 4 while the number of rows in Y is also 4. They will result in a 5x5 matrix.

If two matrices are not compatible because of a shape such as a 5x4 dot 5x4 then we need to transpose one of them to get them compatible.

## Without transposition

```python
X = np.array([[0.49, 0.97, 0.53, 0.05],
          [0.33, 0.65, 0.62, 0.51],
          [1, 0.38, 0.61, 0.45],
          [0.74, 0.27, 0.64, 0.17],
          [0.36, 0.17, 0.96, 0.12]])

print(X.shape)

Y = np.array([[0.79, 0.32, 0.68, 0.9, 0.77],
           [0.18, 0.39, 0.12, 0.93, 0.09],
           [0.87, 0.42, 0.6, 0.71, 0.12],
           [0.45, 0.55, 0.40, 0.78, 0.81]])

print(Y.shape)

           
matrix_product = np.dot(X,Y).round(2)
print(matrix_product.shape)
matrix_product  
```

The below will result in an error

```python
X = np.array([[0.49, 0.33, 1.  , 0.74, 0.36],
       [0.97, 0.65, 0.38, 0.27, 0.17],
       [0.53, 0.62, 0.61, 0.64, 0.96],
       [0.05, 0.51, 0.45, 0.17, 0.12]])

print(X.shape)

Y = np.array([[0.79, 0.32, 0.68, 0.9, 0.77],
           [0.18, 0.39, 0.12, 0.93, 0.09],
           [0.87, 0.42, 0.6, 0.71, 0.12],
           [0.45, 0.55, 0.40, 0.78, 0.81]])

print(Y.shape)

matrix_product = np.dot(X,Y).round(2)
print(matrix_product.shape)
matrix_product  
```

Transposing X will solve the problem.


```python
X = np.array([[0.49, 0.33, 1.  , 0.74, 0.36],
       [0.97, 0.65, 0.38, 0.27, 0.17],
       [0.53, 0.62, 0.61, 0.64, 0.96],
       [0.05, 0.51, 0.45, 0.17, 0.12]])

X_transposed = X.T

print(X.shape)
print(X_transposed.shape)

Y = np.array([[0.79, 0.32, 0.68, 0.9, 0.77],
           [0.18, 0.39, 0.12, 0.93, 0.09],
           [0.87, 0.42, 0.6, 0.71, 0.12],
           [0.45, 0.55, 0.40, 0.78, 0.81]])

print(Y.shape)

matrix_product = np.dot(X_transposed,Y).round(2)
print(matrix_product.shape)
matrix_product  
```
